```python
# 가해운전자 연령대별 교통사고 현황
import matplotlib.pyplot as plt

age_groups = ["20세이하", "21-30세", "31-40세", "41-50세", "51-60세", "61-64세", "65세이상", "불명"]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

# 파이차트 
plt.figure(figsize=(15, 7))
plt.subplot(1, 2, 1)
plt.pie(accident_counts, labels=age_groups, autopct='%1.1f%%', startangle=90)
plt.title('가해운전자 연령대별 교통사고 건수 비율')

# 바차트 
plt.subplot(1, 2, 2)
bar_width = 0.3
index = range(len(age_groups))
plt.bar(index, death_counts, width=bar_width, label='사망자수')
plt.bar([i + bar_width for i in index], injury_counts, width=bar_width, label='부상자수')
plt.xlabel('연령대')
plt.ylabel('인원수')
plt.title('가해운전자 연령대별 교통사고 사망자 및 부상자 현황')
plt.xticks([i + bar_width/2 for i in index], age_groups)
plt.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```
