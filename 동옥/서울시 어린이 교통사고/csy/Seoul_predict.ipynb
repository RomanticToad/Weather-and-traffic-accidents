```python
# Radar Chart 
# 어린이 사고 수 비교 ( 종로구 vs 강남구 ) - 변수 간의 상대적인 크기, 패턴 강조

import numpy as np

categories = ['어린이 사고 수', '어린이보호구역 CCTV 수', '어린이 공원 개수', '어린이 집 수', '유치원 수']
values_jongro = [28, 48, 24, 116, 17]
values_gangnam = [115, 1157, 62, 359, 40]

# 첫 번째 값 추가 - 원형 생성
categories += [categories[0]]
values_jongro += [values_jongro[0]]
values_gangnam += [values_gangnam[0]]

# 각 변수의 각도 계산
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False)

# 그래프 그리기
plt.figure(figsize=(8, 8))
plt.polar(angles, values_jongro, marker='o', linestyle='solid', linewidth=2, color='skyblue', fill=True, label='종로구')
plt.polar(angles, values_gangnam, marker='o', linestyle='solid', linewidth=2, color='lightcoral', fill=True, label='강남구')

# 각 변수에 라벨 달기
plt.xticks(angles[:-1], categories, color='grey', size=12)

# 그래프 제목
plt.title('종로구 vs 강남구', size=18, color='blue', y=1.1)

# 범례 추가
plt.legend(loc='upper right')

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib import rc  
rc('font', family='AppleGothic') 			
plt.rcParams['axes.unicode_minus'] = False  

# 주어진 데이터
data = {
    '지역구': ['종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구', '강서구', '구로구', '금천구', '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구'],
    '어린이사고수': [28, 19, 36, 38, 72, 60, 95, 82, 59, 54, 92, 85, 44, 57, 114, 99, 94, 43, 88, 66, 58, 80, 115, 140, 87]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 그래프 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
plt.rcParams['hatch.linewidth'] = 3

# 어린이 사고 수에 따라 hatch 패턴을 다르게 적용한 막대 그래프
bars = plt.bar(df['지역구'], df['어린이사고수'], facecolor='skyblue', edgecolor='black')

for i, bar in enumerate(bars):
    hatch_str = "/" * df['어린이사고수'].iloc[i]
    bar.set_hatch(hatch_str)

plt.title('어린이 사고 수에 따른 막대 그래프')
plt.xlabel('지역구')
plt.ylabel('어린이 사고 수')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
