```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
# 데이터 불러오기 (가정)
# data = pd.read_csv('your_data.csv')
# 실제 데이터 파일 경로를 지정해주세요.

# CSV 파일 경로 지정
csv_file_path = '교통사고.csv'

# 데이터 불러오기
data = pd.read_csv(csv_file_path)


# 1. 사고 유형 분석
sns.countplot(x='사고유형', data=data)
plt.title('사고 유형별 발생 건수')
plt.xlabel('사고 유형')
plt.ylabel('발생 건수')
plt.xticks(rotation=45)
plt.tight_layout()  # 여기에 추가
plt.show()

# 1. 사고 부문과 사고 내용을 결합하여 새로운 열 생성
data['사고 부문 및 내용'] = data['사고부문'] + ' - ' + data['사고내용']

# 1-1. 사고 부문과 사고 내용 조합별 발생 건수 시각화
plt.figure(figsize=(15, 6))
sns.countplot(x='사고 부문 및 내용', data=data, order=data['사고 부문 및 내용'].value_counts().index)
plt.title('사고 부문 및 내용별 발생 건수')
plt.xlabel('사고 부문 및 내용')
plt.ylabel('발생 건수')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


# 2. 사고 일시 및 시간대 분석
data['사고일시'] = pd.to_datetime(data['사고일시'], format='%Y년 %m월 %d일 %H시')

data['요일'] = pd.Categorical(data['요일'], categories=['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일'], ordered=True)
sns.countplot(x='요일', data=data, order=['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일'])
plt.title('요일별 발생 건수')
plt.xlabel('요일')
plt.ylabel('발생 건수')
plt.show()

# 3. 지역별 사고 분포 시각화 (Top 15)
top15_regions = data['시군구'].value_counts().nlargest(15)  # Top 15 지역 선택
plt.figure(figsize=(12, 6))
sns.countplot(x='시군구', data=data, order=top15_regions.index)
plt.title('지역별 발생 건수 (Top 15)')
plt.xlabel('시군구')
plt.ylabel('발생 건수')

# 발생 건수 숫자로 표시
for i, v in enumerate(top15_regions.values):
    plt.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=8, color='black')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 각 지역별 사고 건수 확인
accidents_by_region = data['시군구'].value_counts()

# 결과 출력
print("각 지역별 사고 건수:")
print(accidents_by_region)
# 총 사고 건수 확인
total_accidents = len(data)

# 결과 출력
print("총 사고 건수:", total_accidents)

# 4. 가해운전자 연령 분석
# '가해운전자 연령' 열에서 숫자만 추출하여 '가해운전자 연령_숫자' 열 생성
data['가해운전자 연령_숫자'] = data['가해운전자 연령'].apply(lambda x: int(re.search(r'\d+', str(x)).group()) if re.search(r'\d+', str(x)) else None)

# '가해운전자 연령_숫자'를 10단위로 나누어 '가해운전자 연령대' 열 생성
data['가해운전자 연령대'] = pd.cut(data['가해운전자 연령_숫자'], bins=range(10, 101, 10), right=False, labels=[f'{i}대' for i in range(10, 100, 10)])

# 시각화
plt.figure(figsize=(12, 6))
sns.countplot(x='가해운전자 연령대', data=data, order=data['가해운전자 연령대'].value_counts().index)
plt.title('가해운전자 연령대별 발생 건수')
plt.xlabel('가해운전자 연령대')
plt.ylabel('발생 건수')
plt.show()

# 5. 사고유형 및 법규위반 연관성 분석
plt.figure(figsize=(12, 6))
sns.countplot(x='사고유형', hue='법규위반', data=data)
plt.title('사고유형 및 법규위반 연관성')
plt.xlabel('사고 유형')
plt.ylabel('발생 건수')
plt.legend(title='법규위반', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()

# 6. 상해 정도에 따른 분석
sns.countplot(x='가해운전자 상해정도', data=data, label='가해운전자', alpha=0.7)
sns.countplot(x='피해운전자 상해정도', data=data, label='피해운전자', alpha=0.7)
plt.title('상해정도별 발생 건수')
plt.xlabel('상해정도')
plt.ylabel('발생 건수')
plt.legend()
plt.show()

# 7. 날씨 및 도로 상태 분석
plt.figure(figsize=(12, 6))
sns.countplot(x='노면상태', hue='기상상태', data=data)
plt.title('날씨 및 도로 상태별 발생 건수')
plt.xlabel('노면상태')
plt.ylabel('발생 건수')
plt.legend(title='기상상태', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()

# 8. 가해운전자 성별에 따른 분석
sns.countplot(x='가해운전자 성별', data=data)
plt.title('가해운전자 성별별 발생 건수')
plt.xlabel('가해운전자 성별')
plt.ylabel('발생 건수')
plt.show()
# 9. 지역별 사고 부문 분석 (가장 많이 발생한 지역 Top 1)
top1_region = data['시군구'].value_counts().idxmax()  # 가장 많이 발생한 지역 (Top 1)
top1_region_data = data[data['시군구'] == top1_region]  # 해당 지역의 데이터 추출

plt.figure(figsize=(15, 6))
sns.countplot(x='사고 부문 및 내용', data=top1_region_data, order=top1_region_data['사고 부문 및 내용'].value_counts().index)
plt.title(f'{top1_region} 지역의 사고 부문 및 내용별 발생 건수')
plt.xlabel('사고 부문 및 내용')
plt.ylabel('발생 건수')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
