```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 한글 폰트 파일 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# CSV 파일 로드
file_path_accident = '서울 21,22 음주운전사고현황.csv'
file_path_traffic = '서울 21,22 교통사고현황.csv'

df_accident = pd.read_csv(file_path_accident, encoding='cp949')
df_traffic = pd.read_csv(file_path_traffic, encoding='cp949')

# X축과 Y축 선택
x_column = '지역구'
y_column_accident_2021 = '2021년'
y_column_accident_2022 = '2022년'
y_column_traffic_2021 = '2021년'
y_column_traffic_2022 = '2022년'

# 음주운전사고 증감율 계산
df_accident['증감율'] = ((df_accident[y_column_accident_2022] - df_accident[y_column_accident_2021]) / df_accident[y_column_accident_2021]) * 100

# 교통사고 증감율 계산
df_traffic['증감율'] = ((df_traffic[y_column_traffic_2022] - df_traffic[y_column_traffic_2021]) / df_traffic[y_column_traffic_2021]) * 100

# 음주운전사고와 교통사고 증감율의 차이 계산
df_difference = pd.DataFrame({
    '지역구': df_accident[x_column],
    '차이': abs(df_accident['증감율'] - df_traffic['증감율'])
})

# 차이를 기준으로 내림차순 정렬
df_difference = df_difference.sort_values(by='차이', ascending=False)

# 그래프 그리기
plt.figure(figsize=(15, 8))

# 음주운전사고와 교통사고 증감율의 차이 그래프
bar_colors = ['lightblue' if diff < 0 else 'lightcoral' for diff in df_difference['차이']]
plt.bar(df_difference['지역구'], df_difference['차이'], color=bar_colors, label='차이')
plt.title('서울 지역구 음주운전사고 및 교통사고 증감율 차이 (2021-2022)')
plt.xlabel('지역구')
plt.ylabel('증감율 차이 절대값 (%)')
plt.xticks(rotation=45)

# 범례 추가
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일 로드
file_path_cctv = '서울 면적cctv.csv'
file_path_traffic = '서울 21,22 교통사고현황.csv'

df_cctv = pd.read_csv(file_path_cctv, encoding='cp949')
df_traffic = pd.read_csv(file_path_traffic, encoding='cp949')

# 두 데이터프레임을 '자치구' 기준으로 병합
df_merged = pd.merge(df_cctv[['지역구', '면적당cctv']], df_traffic[['지역구', '2022년']], left_on='지역구', right_on='지역구', how='inner')

# 산점도 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(x='면적당cctv', y='2022년', data=df_merged, hue='지역구', palette='viridis', s=100)

plt.title('단위면적당 CCTV 갯수와 2022년 교통사고 건수의 상관관계')
plt.xlabel('면적당 CCTV 갯수')
plt.ylabel('2022년 교통사고 건수')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
