```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 한글 폰트 경로 (Windows 기준)
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# CSV 파일 경로
file_path = '2017~2019 서울시 교통사고.csv'

# CSV 파일 읽기 (encoding 지정)
df = pd.read_csv(file_path, delimiter=',', encoding='cp949')

# '사고일시' 컬럼을 datetime 형식으로 변환
df['사고일시'] = pd.to_datetime(df['사고일시'], format='%Y년 %m월 %d일 %H시')

# '월' 컬럼 추가
df['월'] = df['사고일시'].dt.month

# 월별 사고 수 계산
monthly_accidents = df.groupby('월').size()

# 꺾은선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(monthly_accidents.index, monthly_accidents.values, marker='o', color='skyblue', linestyle='-', linewidth=2)
plt.title('월별 사고 수')
plt.xlabel('월')
plt.ylabel('사고 수')
plt.xticks(monthly_accidents.index)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 한글 폰트 경로 (Windows 기준)
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# CSV 파일 경로
file_path_accident = '2017~2019 서울시 교통사고.csv'
file_path_child_accident = '2017~2019 서울시 어린이 보행자사고.csv'

# CSV 파일 읽기 (encoding 지정)
df_accident = pd.read_csv(file_path_accident, delimiter=',', encoding='cp949')
df_child_accident = pd.read_csv(file_path_child_accident, delimiter=',', encoding='cp949')

# '시군구' 컬럼에서 첫 번째 영역 추출하여 새로운 컬럼 생성
df_accident['지역구'] = df_accident['시군구'].apply(lambda x: x.split()[1] if pd.notnull(x) and isinstance(x, str) else None)
df_child_accident['지역구'] = df_child_accident['시군구'].apply(lambda x: x.split()[1] if pd.notnull(x) and isinstance(x, str) else None)

# '미분류'인 데이터 제외
df_accident = df_accident[df_accident['지역구'] != '미분류']
df_child_accident = df_child_accident[df_child_accident['지역구'] != '미분류']

# '지역구' 컬럼을 기준으로 그룹화하여 전체 사고 수 및 어린이 사고 수 계산
전체_사고수 = df_accident.groupby('지역구').size().sort_index()
어린이_사고수 = df_child_accident.groupby('지역구').size().sort_index()

# 전체 사고 대비 어린이 사고 비율 계산
어린이_비율 = 어린이_사고수 / 전체_사고수

# 막대 그래프 그리기
plt.figure(figsize=(15, 6))
어린이_비율.plot(kind='bar', color=['skyblue','skyblue','skyblue','skyblue','skyblue','skyblue','skyblue','skyblue','skyblue','skyblue','skyblue','skyblue',
                                 'skyblue','skyblue','skyblue','skyblue','skyblue','skyblue','lightcoral','skyblue','skyblue','lightcoral','grey','grey','skyblue'])
plt.title('지역별 전체 교통사고 대비 어린이 보행자 사고 비율 (미분류 제외)')
plt.xlabel('지역구')
plt.ylabel('어린이 보행자 사고 비율')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 한글 폰트 경로 (Windows 기준)
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# CSV 파일 경로
file_path = '2017~2019 서울시 교통사고.csv'

# CSV 파일 읽기 (encoding 지정)
df = pd.read_csv(file_path, delimiter=',', encoding='cp949')

# '법규위반' 컬럼을 기준으로 그룹화하여 사고 수 계산
전체_사고수 = df.groupby('법규위반').size().sort_values(ascending=False)

# 전체 사고 대비 법규위반별 사고 비율 계산
사고_비율 = (전체_사고수 / df.shape[0]).sort_values(ascending=False)

# 꺾은선 그래프 그리기
plt.figure(figsize=(15, 6))
plt.plot(사고_비율.index, 사고_비율.values, marker='o', color='skyblue', linestyle='-', linewidth=2)
plt.title('전체 교통사고 기준 법규위반별 사고 비율')
plt.xlabel('법규위반')
plt.ylabel('사고 비율')
plt.xticks(rotation=45, ha='right')
plt.grid(False)  # 배경 그리드 비활성화
plt.show()
```

```python
# '법규위반' 컬럼을 기준으로 그룹화하여 사고 수 계산
전체_사고수 = df1.groupby('법규위반').size().sort_values(ascending=False)
어린이_사고수 = df2.groupby('법규위반').size().sort_values(ascending=False)

# 전체 사고 대비 법규위반별 사고 비율 계산
사고_비율1 = (전체_사고수 / df1.shape[0]).sort_values(ascending=False)
사고_비율2 = (어린이_사고수 / df2.shape[0]).sort_values(ascending=False)

# 꺾은선 그래프 그리기
plt.figure(figsize=(15, 6))

# 그래프 그리기 (파란색) - 전체 교통사고
plt.plot(어린이_사고수.index, 사고_비율1.reindex(어린이_사고수.index).values, marker='o', color='skyblue', linestyle='-', linewidth=2, label='전체 교통사고')

# 그래프 그리기 (분홍색) - 어린이 보행자 사고
plt.plot(어린이_사고수.index, 사고_비율2.values, marker='o', color='lightcoral', linestyle='-', linewidth=2, label='어린이 보행자 사고')

plt.title('전체 교통사고 및 어린이 보행자 사고 기준 법규위반별 사고 비율')
plt.xlabel('법규위반')
plt.ylabel('사고 비율')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(False)
plt.show()
```
