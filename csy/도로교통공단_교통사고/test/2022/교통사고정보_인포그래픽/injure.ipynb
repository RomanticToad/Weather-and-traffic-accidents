```python
import matplotlib.pyplot as plt

# 데이터 정의
data = {
    "사고시상태": ["횡단보도내", "횡단보도외", "마주보고 통행", "등지고 통행", "보도 통행", "길가장자리구역 통행",
                "승차 중 관련", "하차 중 관련", "도로위 작업 중", "도로위 놀이 중", "기타"],
    "사망자수": [243, 266, 51, 78, 14, 36, 2, 3, 24, 1, 215],
    "부상자수": [10788, 5898, 1762, 2787, 2409, 2508, 458, 412, 565, 144, 10357]
}

# 그래프 생성
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# 사망자수 파이 차트
axes[0].pie(data["사망자수"], labels=data["사고시상태"], autopct='%1.1f%%', startangle=90)
axes[0].set_title('사망자수')

# 부상자수 파이 차트
axes[1].pie(data["부상자수"], labels=data["사고시상태"], autopct='%1.1f%%', startangle=90)
axes[1].set_title('부상자수')

# 중상자수 파이 차트
axes[2].pie(data["부상자수"], labels=data["사고시상태"], autopct='%1.1f%%', startangle=90)
axes[2].set_title('중상자수')

# 그래프 표시
plt.show()
```
