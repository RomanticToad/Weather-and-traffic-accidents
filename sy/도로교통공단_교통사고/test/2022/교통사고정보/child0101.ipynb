```python
import matplotlib.pyplot as plt

data = {
    "사고시상태": ["자동차 승차중", "이륜차 승차중", "자전거 승차중", "보행중", "기타"],
    "사망자수": [3, 0, 1, 14, 0],
    "부상자수": [7367, 38, 1106, 2838, 40]
}

# pie 함수 -> 사망자수와 부상자수에 대한 파이 차트 생성
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6)) # 그래프 생성

# 파이 차트 - 사망자 수 
axes[0].pie(data["사망자수"], labels=data["사고시상태"], autopct='%1.1f%%', startangle=90)
axes[0].set_title('사망자수')

# 파이 차트 - 부상자 수 
axes[1].pie(data["부상자수"], labels=data["사고시상태"], autopct='%1.1f%%', startangle=90)
axes[1].set_title('부상자수')

plt.show()
```
