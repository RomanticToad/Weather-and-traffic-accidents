```python
import matplotlib.pyplot as plt

# 데이터 정의
data = {
    "시도": ["서울", "부산", "대구", "인천", "광주", "대전", "울산", "세종", "경기", "강원", "충북", "충남", "전북", "전남", "경북", "경남", "제주"],
    "교통안전지수": [79.09, 78.43, 76.51, 79.07, 79.28, 78.16, 82.64, 76.65, 68.12, 79.79, 64.24, 67.96, 74.17, 74.16, 74.09, 75.24, 69.08],
    "순위": [35, 42, 57, 36, 34, 46, 7, 6, 24, 3, 28, 25, 12, 47, 13, 9, 23]
}

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(15, 7))

# 교통안전지수 선 그래프
ax1.plot(data["시도"], data["교통안전지수"], marker='o', color='skyblue', label='교통안전지수')
ax1.set_xlabel('시도')
ax1.set_ylabel('교통안전지수', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_title('교통안전지수 및 순위')

# 순위 선 그래프 (축을 오른쪽에 추가)
ax2 = ax1.twinx()
ax2.plot(data["시도"], data["순위"], marker='o', color='salmon', label='순위')
ax2.set_ylabel('순위', color='salmon')
ax2.tick_params(axis='y', labelcolor='salmon')

# 그래프 표시
plt.show()
```

```python
# 산점도
import matplotlib.pyplot as plt

# 데이터 정의
data = {
    "시도": ["서울", "부산", "대구", "인천", "광주", "대전", "울산", "세종", "경기", "강원", "충북", "충남", "전북", "전남", "경북", "경남", "제주"],
    "교통안전지수": [79.09, 78.43, 76.51, 79.07, 79.28, 78.16, 82.64, 76.65, 68.12, 79.79, 64.24, 67.96, 74.17, 74.16, 74.09, 75.24, 69.08],
    "순위": [35, 42, 57, 36, 34, 46, 7, 6, 24, 3, 28, 25, 12, 47, 13, 9, 23]
}

# 그래프 생성
fig, ax = plt.subplots(figsize=(10, 7))

# 산점도 그래프
scatter = ax.scatter(data["교통안전지수"], data["순위"], c='skyblue', alpha=0.7, edgecolors='w')

# 각 점에 대한 레이블 표시
for i, txt in enumerate(data["시도"]):
    ax.annotate(txt, (data["교통안전지수"][i], data["순위"][i]), fontsize=9, ha='right')

# 축 및 그래프 제목 설정
ax.set_xlabel('교통안전지수')
ax.set_ylabel('순위')
ax.set_title('교통안전지수와 순위의 산점도 그래프')

# 그래프 표시
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 정의
data = {
    "시도": ["서울", "부산", "대구", "인천", "광주", "대전", "울산", "세종", "경기", "강원", "충북", "충남", "전북", "전남", "경북", "경남", "제주"],
    "교통안전지수": [79.09, 78.43, 76.51, 79.07, 79.28, 78.16, 82.64, 76.65, 68.12, 79.79, 64.24, 67.96, 74.17, 74.16, 74.09, 75.24, 69.08],
    "순위": [35, 42, 57, 36, 34, 46, 7, 6, 24, 3, 28, 25, 12, 47, 13, 9, 23]
}

# 그래프 생성
plt.figure(figsize=(12, 6))

# 밀도 플롯 생성
sns.kdeplot(data=data["교통안전지수"], label='교통안전지수', fill=True)
sns.kdeplot(data=data["순위"], label='순위', fill=True)

# 그래프 제목 설정
plt.title('교통안전지수와 순위의 밀도 플롯')

# 그래프 표시
plt.show()
```
