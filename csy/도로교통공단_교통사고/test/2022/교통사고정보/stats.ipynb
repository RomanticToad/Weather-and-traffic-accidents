```python
# 버블 차트 - 3개의 변수를 동시에 표현
# 부상자 수 = 버블 크기, 사망자 수 = 빨간색, 교통사고 건수 = 파란색

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc 
rc('font', family='AppleGothic') 		
plt.rcParams['axes.unicode_minus'] = False  


age_groups = ["20세이하", "21-30세", "31-40세", "41-50세", "51-60세", "61-64세", "65세이상", "불명"]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.scatter(age_groups, accident_counts, s=np.array(injury_counts) / 100, c='blue', alpha=0.5, label='부상자 수')
plt.scatter(age_groups, death_counts, s=np.array(death_counts) * 2, c='red', alpha=0.5, label='사망자 수')

plt.title('가해운전자 연령대별 교통사고 현황 (버블 차트)')
plt.xlabel('연령대')
plt.ylabel('교통사고 건수')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 정의
age_groups = ["20세이하", "21-30세", "31-40세", "41-50세", "51-60세", "61-64세", "65세이상", "불명"]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

# 데이터 프레임 생성
df = pd.DataFrame({
    '연령대': age_groups * 3,
    '사고유형': ['사고건수'] * 8 + ['사망자수'] * 8 + ['부상자수'] * 8,
    '인원수': accident_counts + death_counts + injury_counts
})

# 박스 플롯
plt.figure(figsize=(12, 6))
sns.boxplot(x='연령대', y='인원수', hue='사고유형', data=df)
plt.title('가해운전자 연령대별 교통사고 현황 - 박스 플롯')
plt.xlabel('연령대')
plt.ylabel('인원수')
plt.show()


# 선 그래프
plt.figure(figsize=(12, 6))
sns.lineplot(x='연령대', y='인원수', hue='사고유형', data=df)
plt.title('가해운전자 연령대별 교통사고 현황 - 선 그래프')
plt.xlabel('연령대')
plt.ylabel('인원수')
plt.show()

# 막대 그래프
plt.figure(figsize=(12, 6))
sns.barplot(x='연령대', y='인원수', hue='사고유형', data=df)
plt.title('가해운전자 연령대별 교통사고 현황 - 막대 그래프')
plt.xlabel('연령대')
plt.ylabel('인원수')
plt.show()

```
